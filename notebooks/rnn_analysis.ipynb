```python
import matplotlib.pyplot as plt

from models.rnn import HistoneRNN
```

```python
model = HistoneRNN(loss='mean_squared_error', n_epochs=30)
y_pred, y_true = model.fit_predict()

model2 = HistoneRNN(loss='huber_loss', n_epochs=30)
y_pred2, y_true2 = model2.fit_predict()
```

```python
score1 = HistoneRNN.get_score(y_pred, y_true)
score2 = HistoneRNN.get_score(y_pred2, y_true2)


fig = plt.figure()
ax = plt.gca()
ax.set_ylabel('True gene expression (log-scale)')
ax.set_xlabel('Predicted gene expression')
blue = ax.scatter(x=y_pred.reshape((y_pred.shape[0])), y=y_true.reshape((y_true.shape[0])), alpha=0.5, s=1, color='blue', label=f'MSE loss, SCC {score1:.3f}')
red = ax.scatter(x=y_pred2.reshape((y_pred2.shape[0])), y=y_true2.reshape((y_true2.shape[0])), alpha=0.5, s=3, color='red', label=f'Huber loss, SCC {score2:.3f}')
ax.legend(handles=[blue, red])
ax.set_yscale('log')

plt.show()
```
